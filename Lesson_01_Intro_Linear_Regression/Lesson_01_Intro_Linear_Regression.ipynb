Отрисуем график функции $y = x^2$

````python
# создадим функцию, возвращающую квадрат числа
def x2(x): return x*x
def x4(x): return x*x*x*x

# Обозначим иксы в промежутке от -4 до 4
xs = list(range(-4, 5))
xs2 = list(range(-2, 3))

# рассчитаем значения y
ys = [ x2(x) for x in xs ]
ys2 = [ x4(x) for x in xs2 ]

# отрисовывать можно и просто вызвав функцию plot
plt.plot(xs, ys)
plt.plot(xs2, ys2)

plt.show()
````

````python
# вместо линий можно рисовать точки
plt.scatter(xs, ys)
plt.scatter(xs2, ys2)
````

````python
# Так же можно накладывать графики друг на друга
import math

# добавим график синуса
sin_ys = [ math.sin(x) for x in xs ]
cos_ys = [ math.cos(x) for x in xs ]

plt.plot(xs, ys) # синяя линяя
plt.scatter(xs, ys) # синие точки

plt.plot(xs, sin_ys) # оранжевая линяя
plt.scatter(xs, sin_ys) # оранжевые точки

plt.plot(xs2, ys2) # зеленая линяя
plt.scatter(xs2, ys2) # зеленые точки

plt.plot(xs, cos_ys, color = "red") # красная линяя
plt.scatter(xs, cos_ys) # красные точки
````

Перемножение матриц


````python
# Создадим по две матрицы размером 100 на 100 на питоне и в numpy

n_rows = 100
n_cols = 100

python_A = [ list(range(n_cols)) for _ in range(n_rows) ]
python_B = [ list(range(n_cols)) for _ in range(n_rows) ]

np_A = np.array([ np.array(r) for r in python_A ])
np_B = np.array([ np.array(r) for r in python_B ])
````

````python
# Медленная кастомная функция перемножения двух матриц (не надо так)
def custom_mat_mul(A, B):
  n = len(A)
  k1 = len(A[0])
  k2 = len(B)
  m = len(B[0])

  if k1 != k2:
    raise ValueError(f'Multiplying matrix with shape {n}x{k1} and matrix with shape {k1}x{m}. {k1} != {k2}')

  k = k1

  ans = [ [0] * n for _ in range(m) ]
  for t in range(m):
    for i in range(n):
      v = 0
      for j in range(k):
        v += A[i][j] * B[j][t]
      ans[i][t] = v

  return ans
````

````python
# Проверим, что ответ нашей функции совпадает с numpy
assert (np_A @ np_B == custom_mat_mul(python_A, python_B)).all() == True

# операнд @ озачает матричное перемножение, также явно можно использовать np.matmul
assert (np.matmul(np_A, np_B) == np_A @ np_B).all() == True
````

````python
%timeit -n 25 custom_mat_mul(python_A, python_B)
````

````python
%timeit -n 25 np_A @ np_B
````

В этом случае разница уже в **200** раз.


### Почему numpy быстрый

В отличие от питоновских листов, которые могут содержать разнородные объекты - из-за чего они медленные, ndarray массивы numpy однородны, то есть способны содержать в себе только данные одного типа, фиксированной длины. Это позволяет хранить все данные **последовательно** в памяти.

Это способствует лучшему кэшированию CPU - что сильно ускоряет вычисления.

Так же numpy разбивает большие операции на множество маленьких и параллелизирует их с помощью SIMD (SSE, AVX).

Но за всё приходится платить, и однородность данных накладывает свои ограничения - операция добавления элемента **очень дорогая**, она требует полной реаллокации массива.

Поэтому *рекомендуется* проводить все манипуляции по расширению массива в питоновском листе, а затем конвертировать его в ndarray

````python
def plot_target_in_bins(df, target_col, col_to_bin, q=10): # q - интервалы на которые будут разбиты данные
  df = df.copy()
  df['bin'] = pd.qcut(df[col_to_bin], q=q, duplicates='drop')

  agg = df.groupby('bin')[[col_to_bin, target_col]].mean() # среднее значение SalePrice в каждом бине
  display(agg)
  plt.plot(agg[col_to_bin], agg[target_col])
````

````python
plt.figure(figsize=(16,9))
plot_target_in_bins(df, 'SalePrice', 'GrLivArea')
````

## Одномерный метод наименьших квадратов

````python
def least_squared_estimator(X, y, learn_intercept=False):
  X = np.array(X)
  if learn_intercept:
    X = np.concatenate([np.ones((X.shape[0], 1), ), X ], axis=1)
  coefs = np.linalg.inv(X.T @ X) @ X.T @ y
  return coefs
````

### 1) corrcoef
Реализовать функцию расчета [коэффицента корреляции Пирсона](https://ru.wikipedia.org/wiki/%D0%9A%D0%BE%D1%80%D1%80%D0%B5%D0%BB%D1%8F%D1%86%D0%B8%D1%8F#%D0%9B%D0%B8%D0%BD%D0%B5%D0%B9%D0%BD%D1%8B%D0%B9_%D0%BA%D0%BE%D1%8D%D1%84%D1%84%D0%B8%D1%86%D0%B8%D0%B5%D0%BD%D1%82_%D0%BA%D0%BE%D1%80%D1%80%D0%B5%D0%BB%D1%8F%D1%86%D0%B8%D0%B8) для двух numpy векторов с помощью функционала numpy (без циклов)

P.S. нельзя использовать функцию np.corrcoef

````python
def corrcoef(v1, v2):
  assert v1.shape == v2.shape
  average_value_v1 = np.mean(v1)
  average_value_v2 = np.mean(v2)
  
  dev_v1 = v1 - average_value_v1
  dev_v2 = v2 - average_value_v2
  
  sum_dev = np.sum(dev_v1 * dev_v2)
  mult_dev = np.sqrt(np.sum(dev_v1 ** 2)) * (np.sqrt(np.sum(dev_v2 ** 2)))
  
  if mult_dev == 0:
    return 0
  else: 
    return sum_dev / mult_dev
  
  pass
````

````python
# Test 1
v1 = np.random.rand(100)
v2 = np.random.rand(100)
assert np.corrcoef(v1, v2)[0][1].round(8) == corrcoef(v1, v2).round(8)
````

````python
# Test 2
v1 = np.array([1, 2, 3])
v2 = np.array([1, 2, 3])
assert np.corrcoef(v1, v2)[0][1].round(8) == corrcoef(v1, v2).round(8)
````

````python
# Test 3
v1 = np.random.normal(loc=0, scale=1, size=100)
v2 = v1 + 0.2 * np.random.rand(100)
assert np.corrcoef(v1, v2)[0][1].round(8) == corrcoef(v1, v2).round(8)
````

### 2) isna
Реализовать функцию, которая принимает произвольный ndarray и возвращает ndarray с таким же shape со значениями True или False в зависимости был ли np.nan в этом месте в оригинальном ndarray

Например: для массива isna(np.array([1, 2, np.nan, 3])) -> np.array([False, False, True, False])

````python
def isna(x):
    y = np.zeros(x.shape, dtype=bool)
    
    for i in range(x.shape[0]):
      if len(x.shape) > 1:
        for j in range(x.shape[1]):
          if np.isnan(x[i, j]):
            y[i, j] = True
          else:
            y[i, j] = False 
      else:
        if np.isnan(x[i]):
          y[i] = True
        else:
          y[i] = False
    return y
    pass  
````

````python
# Test 1

a = np.ones(10)
a[0] = np.nan
a[1] = None

assert (isna(a) == np.isnan(a)).all()
````

````python
# Test 2

a = np.ones((10, 10))
a[0][5] = np.nan
a[9][2] = np.nan

assert (isna(a) == np.isnan(a)).all()
````

### 3) dist
Реализовать функцию dist(v1, v2), которая принимает на вход два вектора с одинаковым shape, и возвращает число - [евклидово расстояние](https://ru.wikipedia.org/wiki/Евклидова_метрика) между ними.

Так же нельзя использовать циклы и np.linagl.norm

````python
def dist(v1, v2):
  if v1.shape != v2.shape:
    return 0
      
  return np.sqrt(np.sum((v1 - v2)**2))
````

````python
# Test 1
v1 = np.array([0, 0, 1])
v2 = np.array([0, 0, 0])

assert dist(v1, v2).round(8) == np.linalg.norm(v1-v2).round(8)
````

````python
# Test 2
v1 = np.array([2, 3, 1])
v2 = np.array([0, 0, 0])

assert dist(v1, v2).round(8) == np.linalg.norm(v1-v2).round(8)
````

````python

from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split

target_col = 'SalePrice'
X = df.drop(columns=[target_col])
y = df[target_col]

outlier_detector = IsolationForest(contamination=0.05, random_state=42)
y_outliers = outlier_detector.fit_predict(y.values.reshape(-1, 1))
outlier_mask = y_outliers == -1

print(f"Обнаружено {sum(outlier_mask)} выбросов из {len(y)} записей")

X_clean = X[~outlier_mask].copy()
y_clean = y[~outlier_mask].copy()

X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.3, random_state=43)

print("Размер тренировочной выборки:", X_train.shape)
print("Размер тестовой выборки:", X_test.shape)
````

````python
missing_values = X_train.isnull().sum()
print("\nКоличество пропущенных значений в каждой колонке:")
print(missing_values[missing_values > 0].sort_values(ascending=False))

print("\nТипы данных колонок:")
print(X_train.dtypes.value_counts())
````

````python
numerical_features = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X_train.select_dtypes(include=['object']).columns.tolist()

print(f"\nЧисловых признаков: {len(numerical_features)}")
print(f"Категориальных признаков: {len(categorical_features)}")
````

````python
correlation_with_target = pd.DataFrame(X_train[numerical_features].corrwith(y_train).abs().sort_values(ascending=False))
correlation_with_target.columns = ['Корреляция с SalePrice']
print("\nТоп-15 числовых признаков по абсолютной корреляции с целевой переменной:")
print(correlation_with_target.head(15))
````

````python
correlation_matrix = X_train[numerical_features].corr().abs()
upper_tri = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))
high_correlation = [(col1, col2, upper_tri.loc[col1, col2]) 
                    for col1 in upper_tri.columns 
                    for col2 in upper_tri.columns 
                    if upper_tri.loc[col1, col2] > 0.8]

print("\nПары признаков с высокой корреляцией (> 0.8):")

for col1, col2, corr in sorted(high_correlation, key=lambda x: x[2], reverse=True):
    print(f"{col1} - {col2}: {corr:.3f}")
````

````python
features_to_drop = []

for col1, col2, _ in high_correlation:
    corr1 = abs(correlation_with_target.loc[col1].values[0])
    corr2 = abs(correlation_with_target.loc[col2].values[0])
    if corr1 >= corr2 and col2 not in features_to_drop:
        features_to_drop.append(col2)
    elif col1 not in features_to_drop:
        features_to_drop.append(col1)

print("\nПризнаки, которые будут удалены из-за высокой корреляции с другими:")
print(features_to_drop)

numerical_features_final = [col for col in numerical_features if col not in features_to_drop]
print(f"\nОставшиеся числовые признаки после удаления высококоррелированных: {len(numerical_features_final)}")
````

````python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Lasso, Ridge

top_n = 5
top_features = correlation_with_target.index[:top_n].tolist()
print(f"\nТоп-{top_n} признаков для нелинейных преобразований:")
print(top_features)

numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

polynomial_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, [col for col in numerical_features_final if col not in top_features]),
        ('poly', polynomial_transformer, top_features),
        ('cat', categorical_transformer, categorical_features)
    ])

lr_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

lr_model.fit(X_train, y_train)
````

````python
#Оптимальное alpha
from sklearn.model_selection import GridSearchCV, KFold

param_grid_ridge = {
    'regressor__alpha': [0.1, 1, 10, 100, 1000]
}

param_grid_lasso = {
    'regressor__alpha': [0.1, 1, 10, 100, 1000],
    'regressor__max_iter': [10000] 
}

ridge_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Ridge())
])

lasso_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Lasso())
])

cv = KFold(n_splits=5, shuffle=True, random_state=42)

grid_search_ridge = GridSearchCV(
    ridge_pipeline, 
    param_grid_ridge, 
    cv=cv, 
    scoring='neg_mean_squared_error',
    n_jobs=-1
)

grid_search_ridge.fit(X_train, y_train)

print("Оптимальные параметры для Ridge:", grid_search_ridge.best_params_)
print("Лучший MSE для Ridge (на кросс-валидации):", -grid_search_ridge.best_score_)

grid_search_lasso = GridSearchCV(
    lasso_pipeline, 
    param_grid_lasso, 
    cv=cv, 
    scoring='neg_mean_squared_error',
    n_jobs=-1
)

grid_search_lasso.fit(X_train, y_train)

print("Оптимальные параметры для Lasso:", grid_search_lasso.best_params_)
print("Лучший MSE для Lasso (на кросс-валидации):", -grid_search_lasso.best_score_)
````

````python
from sklearn.metrics import mean_squared_error, r2_score

y_train_pred_lr = lr_model.predict(X_train)
y_test_pred_lr = lr_model.predict(X_test)

train_mse_lr = mean_squared_error(y_train, y_train_pred_lr)
test_mse_lr = mean_squared_error(y_test, y_test_pred_lr)
train_r2_lr = r2_score(y_train, y_train_pred_lr)
test_r2_lr = r2_score(y_test, y_test_pred_lr)


print(f"MSE на тренировочной выборке: {train_mse_lr:.2f}")
print(f"MSE на тестовой выборке: {test_mse_lr:.2f}")
print(f"R² на тренировочной выборке: {train_r2_lr:.4f}")
print(f"R² на тестовой выборке: {test_r2_lr:.4f}")
````

````python
ridge_alpha = grid_search_ridge.best_params_['regressor__alpha']
ridge_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Ridge(alpha=ridge_alpha))
])

ridge_model.fit(X_train, y_train)

y_train_pred_ridge = ridge_model.predict(X_train)
y_test_pred_ridge = ridge_model.predict(X_test)

train_mse_ridge = mean_squared_error(y_train, y_train_pred_ridge)
test_mse_ridge = mean_squared_error(y_test, y_test_pred_ridge)
train_r2_ridge = r2_score(y_train, y_train_pred_ridge)
test_r2_ridge = r2_score(y_test, y_test_pred_ridge)

print(f"aplha = {ridge_alpha}")
print(f"MSE на тренировочной выборке: {train_mse_ridge:.2f}")
print(f"MSE на тестовой выборке: {test_mse_ridge:.2f}")
print(f"R² на тренировочной выборке: {train_r2_ridge:.4f}")
print(f"R² на тестовой выборке: {test_r2_ridge:.4f}")
````

````python
lasso_alpha = grid_search_lasso.best_params_['regressor__alpha']
lasso_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', Lasso(alpha=lasso_alpha, max_iter=10000))
])

lasso_model.fit(X_train, y_train)

y_train_pred_lasso = lasso_model.predict(X_train)
y_test_pred_lasso = lasso_model.predict(X_test)

train_mse_lasso = mean_squared_error(y_train, y_train_pred_lasso)
test_mse_lasso = mean_squared_error(y_test, y_test_pred_lasso)
train_r2_lasso = r2_score(y_train, y_train_pred_lasso)
test_r2_lasso = r2_score(y_test, y_test_pred_lasso)

print(f"aplha = {lasso_alpha}")
print(f"MSE на тренировочной выборке: {train_mse_lasso:.2f}")
print(f"MSE на тестовой выборке: {test_mse_lasso:.2f}")
print(f"R² на тренировочной выборке: {train_r2_lasso:.4f}")
print(f"R² на тестовой выборке: {test_r2_lasso:.4f}")
````

````python
X_train_transformed = preprocessor.fit_transform(X_train)

num_features_names = [col for col in numerical_features_final if col not in top_features]

poly_transformer = preprocessor.named_transformers_['poly']
poly_features = poly_transformer.named_steps['poly'].get_feature_names_out(top_features)

cat_transformer = preprocessor.named_transformers_['cat']
try:
    cat_features = cat_transformer.named_steps['onehot'].get_feature_names_out(categorical_features)
except:
    cat_features = np.array([f"cat_{i}" for i in range(preprocessor.named_transformers_['cat'].named_steps['onehot'].n_features_out_)])

feature_names = np.concatenate([num_features_names, poly_features, cat_features])

lasso_direct = Lasso(alpha=grid_search_lasso.best_params_['regressor__alpha'], max_iter=10000)
lasso_direct.fit(X_train_transformed, y_train)

coef_df = pd.DataFrame({'Признак': feature_names, 'Коэффициент': lasso_direct.coef_})
coef_df['Абсолютное значение'] = coef_df['Коэффициент'].abs()
coef_df = coef_df.sort_values('Абсолютное значение', ascending=False)

print("\nТоп-15 наиболее важных признаков по Lasso:")
print(coef_df.head(15))

print("\nКоличество признаков с нулевыми коэффициентами (ненужные признаки):")
print(f"Всего признаков: {len(coef_df)}")
print(f"Признаки с ненулевыми коэффициентами: {sum(coef_df['Коэффициент'] != 0)}")
print(f"Признаки с нулевыми коэффициентами: {sum(coef_df['Коэффициент'] == 0)}")
print(f"Процент отсеянных признаков: {sum(coef_df['Коэффициент'] == 0) / len(coef_df) * 100:.2f}%")
````
